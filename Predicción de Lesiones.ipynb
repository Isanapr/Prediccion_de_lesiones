# 2. Imputación de datos

```python
zeros = GPS_filtrado.loc[1,]

def ImputaciónDatos(fecha):
    imputacion = GPS_filtrado.loc[GPS_filtrado['Date'] == fecha]
    missing = set(nombres_unicos).difference(imputacion['Player'])

    for player in missing:
        zeros['Player','Date','Distance(m)','Dist (m/min)','HMLD (m)','% ALTA INTENSIDAD','Distance Acc(m)','Distance Dec(m)',
              'HSR Abs Dist (m)','HSR Abs (m/min)','Sprint Abs(m)','MAX Speed(km/h)','Step Balance(%)','Player Load ',
              'Power met','duration'] = 0
        zeros['Player'] = player
        zeros['Date'] = fecha
        imputacion = imputacion.append(zeros)
        #imputacion.loc[len(imputacion)] = 0

    return imputacion
```

```python
GPS_fulldays = ImputaciónDatos(fechas_unicas[0])
GPS_fulldays['Date'] = fechas_unicas[0]

del fechas_unicas[0]

# Agregarle el resto al dataframe anterior
for fecha in fechas_unicas:
    imputacion = ImputaciónDatos(fecha)
    imputacion['Date'] = fecha
    GPS_fulldays = GPS_fulldays.append(imputacion)
```

# 3. Feature Engineering

### Crear funcion para las nuevas variables

```python
def New_Features(player):
    df_player = GPS_fulldays.loc[GPS_fulldays['Player'] == player]
    df_player = df_player.resample('D', on='Date').sum()
    df_player = df_player.fillna(0)
        
    # Distance
    df_player["Acute_Dis"] = (df_player.loc[:,"Distance(m)"].rolling('7D', 7).sum())/7
    df_player["Chronic_Dis"] = (df_player.loc[:,"Distance(m)"].rolling('28D', 28).sum())/28
    df_player["ACWR_Dis"] = df_player.loc[:,"Acute_Dis"]/df_player.loc[:,"Chronic_Dis"]
    df_player["EWMA_Dis"] = df_player.loc[:,"Distance(m)"].ewm(span=6).mean()
    df_player["Mean_Dis"] = (df_player.loc[:,"Distance(m)"].rolling('7D', 7).sum())/7
    df_player["Std_Dis"] = (df_player.loc[:,"Distance(m)"].rolling('7D', 7).std())
    df_player["MSWR_Dis"] = df_player.loc[:,"Mean_Dis"]/df_player.loc[:,"Std_Dis"]
    
    # HMLD (m)
    df_player["Acute_HMLD"] = (df_player.loc[:,"HMLD (m)"].rolling('7D', 7).sum())/7
    df_player["Chronic_HMLD"] = (df_player.loc[:,"HMLD (m)"].rolling('28D', 28).sum())/28
    df_player["ACWR_HMLD"] = df_player.loc[:,"Acute_HMLD"]/df_player.loc[:,"Chronic_HMLD"]
    df_player["EWMA_HMLD"] = df_player.loc[:,"HMLD (m)"].ewm(span=6).mean()
    df_player["Mean_HMLD"] = (df_player.loc[:,"HMLD (m)"].rolling('7D', 7).sum())/7
    df_player["Std_HMLD"] = (df_player.loc[:,"HMLD (m)"].rolling('7D', 7).std())
    df_player["MSWR_HMLD"] = df_player.loc[:,"Mean_HMLD"]/df_player.loc[:,"Std_HMLD"]
    
     # HSR Abs Dist (m)
    df_player["Acute_HSR"] = (df_player.loc[:,"HSR Abs Dist (m)"].rolling('7D', 7).sum())/7
    df_player["Chronic_HSR"] = (df_player.loc[:,"HSR Abs Dist (m)"].rolling('28D', 28).sum())/28
    df_player["ACWR_HSR"] = df_player.loc[:,"Acute_HSR"]/df_player.loc[:,"Chronic_HSR"]
    df_player["EWMA_HSR"] = df_player.loc[:,"HSR Abs Dist (m)"].ewm(span=6).mean()
    df_player["Mean_HSR"] = (df_player.loc[:,"HSR Abs Dist (m)"].rolling('7D', 7).sum())/7
    df_player["Std_HSR"] = (df_player.loc[:,"HSR Abs Dist (m)"].rolling('7D', 7).std())
    df_player["MSWR_HSR"] = df_player.loc[:,"Mean_HSR"]/df_player.loc[:,"Std_HSR"]
    
    # Power met
    df_player["Acute_PowerM"] = (df_player.loc[:,"Power met"].rolling('7D', 7).sum())/7
    df_player["Chronic_PowerM"] = (df_player.loc[:,"Power met"].rolling('28D', 28).sum())/28
    df_player["ACWR_PowerM"] = df_player.loc[:,"Acute_PowerM"]/df_player.loc[:,"Chronic_PowerM"]
    df_player["EWMA_PowerM"] = df_player.loc[:,"Power met"].ewm(span=6).mean()
    df_player["Mean_PowerM"] = (df_player.loc[:,"Power met"].rolling('7D', 7).sum())/7
    df_player["Std_PowerM"] = (df_player.loc[:,"Power met"].rolling('7D', 7).std())
    df_player["MSWR_PowerM"] = df_player.loc[:,"Mean_PowerM"]/df_player.loc[:,"Std_PowerM"]
    
    return df_player;
```

### Correr función

```python
# Crear nuevo dataframe con el primer nombre
GPS_fe = New_Features(nombres_unicos[0])
GPS_fe['Player'] = nombres_unicos[0]

del nombres_unicos[0]

# Agregarle el resto al dataframe anterior
for player in nombres_unicos:
    df_player = New_Features(player)
    df_player['Player'] = player
    GPS_fe = GPS_fe.append(df_player)
```

```python
datos_personales['Fecha nacimiento'] = datos_personales['Fecha nacimiento'].astype(str)
age = datos_personales['Fecha nacimiento'].str.split("-", n =2, expand = True)
age = age.fillna(-2000)
age = age.astype(int)
age = age.replace('-2000', np.nan)

# Make separate columns in the data frame from training_duration data frame
datos_personales['Edad'] = 2022-age[0]

# Drop old column (i.e., Duration) 
datos_personales.drop(columns = ['Fecha nacimiento'], inplace = True)
```

```python
def AgregarDatos(player):
    
    datos_extras = GPS_fe.loc[GPS_fe['Player'] == player]
    
    peso = datos_personales.loc[player]['Peso']
    talla = datos_personales.loc[player]['Talla']
    rol = datos_personales.loc[player]['Rol']
    edad = datos_personales.loc[player]['Edad']
    
    datos_extras['Peso'] = peso
    datos_extras['Talla'] = talla
    datos_extras['Rol'] = rol
    datos_extras['Edad'] = edad
    
    return datos_extras
```
